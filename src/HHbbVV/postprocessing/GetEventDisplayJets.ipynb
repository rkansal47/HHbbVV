```python
for f in tqdm(files):
    with uproot.open(f"{f}:Runs") as file:
        runs = file["run"].arrays().run.to_numpy()

    runmatch = False
    for run_num in run_nums:
        if np.isin(run_num, runs):
            runmatch = True

    if not runmatch:
        continue

    with uproot.open(f"{f}:LuminosityBlocks") as file:
        lbs = file["luminosityBlock"].arrays().luminosityBlock.to_numpy()

    lumimatch = False
    for lb in lumi_blocks:
        if np.isin(lb, lbs):
            lumimatch = True

    if not lumimatch:
        continue

    print(f"Lumi block match in {f}!")

    with uproot.open(f"{f}:Events/event") as file:
        events = file.arrays().event.to_numpy()

    # print(events)
    # print(f)

    eventmatch = False
    for event_id in event_ids:
        print(event_id)
        if np.isin(event_id, events):
            eventmatch = True
            print(f"Found event {event_id} in {f}!")
            idx = np.where(events == event_id)[0][0]
            print(idx)
            break

    if eventmatch:
        break
```
